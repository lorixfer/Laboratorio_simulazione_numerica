```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math


# Open the file
with open("data.txt", "r") as file:
    # Initialize an empty list to store the numbers from the first column
    x = []
    y = []
    yerr = []
    # Read each line in the file
    for line in file:
        # Split the line into individual numbers
        line_numbers = line.split()

        x.append(float(line_numbers[0]) )
        y.append(float(line_numbers[1]) -0.5 )
        yerr.append(float(line_numbers[2] ) )

# Print the array containing numbers from the first column
print(x)
print(y)
```

```python


plt.errorbar(x, y ,yerr)
plt.xlabel('#throws')
plt.ylabel('<r>-1/2')
plt.grid(True)
plt.show()
```
